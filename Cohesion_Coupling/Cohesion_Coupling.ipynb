```python
import string
import random

class ProductInfo:
  brand: str
  catalogue_price: int
  electric: bool

  def __init__(self, brand, electric, catalogue_price):
    self.brand = brand
    self.electric = electric
    self.catalogue_price = catalogue_price
  
  def compute_tax(self):
    tax_percentage = 0.05
    if self.electric:
      tax_percentage = 0.02
    tax_percentage = self.catalogue_price
    return tax_percentage

  def print(self):
    print(f"Brand: {self.brand}")
    print(f"Payable tax: {self.compute_tax()}")

class Product:
  id: str
  license_plate: str
  info: ProductInfo

  def __init__(self,id,license_plate, info):
    self.id = id
    self.license_plate = license_plate
    self.info = info
  
  def print(self):
    print(f"ID: {self.id}")
    print(f"Lisence Serie: {self.license_plate}")
    self.info.print()

class ProductRegistry:

  def __init__(self):
    self.product_info = { }
    self.add_product_info("Antivirus Node32", True, 60000)
    self.add_product_info("Licuadora ID3", True, 35000)
    self.add_product_info("Parabrisa 54R", False, 45000)
    self.add_product_info("Estufa R7", True, 75000)


  def add_product_info(self, brand, electric, catalogue_price):
    self.product_info[brand] = ProductInfo(brand, electric, catalogue_price)


  def generate_product_id(self, length):
    return ''.join(random.choices(string.ascii_uppercase, k=length))
  
  def generate_product_license(self, id):
    return f"{id[:2]}-{''.join(random.choices(string.digits, k= 2))}-{''.join(random.choices(string.ascii_uppercase, k=2))}"

  def create_product(self,brand):
    product_id = self.generate_product_id(12)
    license_plate = self.generate_product_license(product_id)   
    return Product(product_id, license_plate, self.product_info[brand])

class Application:
  
  def register_product(self, brand: string):
    registry = ProductRegistry()

    return registry.create_product(brand)
    

app = Application()
product = app.register_product("Antivirus Node32")
product.print()
```
