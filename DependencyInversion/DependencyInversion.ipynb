```python
from abc import ABC, abstractmethod


class Switchable(ABC):
    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass


class MotorSwitch(Switchable):
    def turn_on(self):
        print("MotorSwitch: turned the Car on...")

    def turn_off(self):
        print("MotorSwitch: turned the Car off...")


class Driver(Switchable):
    def turn_on(self):
        print("Driver: turned the Car on...")

    def turn_off(self):
        print("Driver: turned the Car off...")


class MechanicalPowerSwitch:

    def __init__(self, driver: Switchable):
        self.client = driver
        self.on = False

    def press(self):
        if self.on:
            self.client.turn_off()
            self.on = False
        else:
            self.client.turn_on()
            self.on = True


l = MotorSwitch()
driver = Driver()
pathSwitch = MechanicalPowerSwitch(driver)
pathSwitch.press()
pathSwitch.press()
```
